## Data visualization - Plots

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../datasets/customer-churn-model/Customer Churn Model.txt")
df.columns.values
```

```python
# Plot scatter of two variables

df.plot(kind="scatter", x="Day Charge", y="Day Mins")
```

```python
# Subplots

figure, axs = plt.subplots(2,2,sharex=True, sharey=True)
df.plot(kind="scatter", x="Day Charge", y="Day Mins", ax=axs[0][0])
```
