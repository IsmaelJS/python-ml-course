## Data visualization - Plots

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("../datasets/customer-churn-model/Customer Churn Model.txt")
df.columns.values
```

```python
# Plot scatter of two variables

df.plot(kind="scatter", x="Day Charge", y="Day Mins")
```

```python
# Subplots

figure, axs = plt.subplots(2,2,sharex=True, sharey=True)
df.plot(kind="scatter", x="Day Charge", y="Day Mins", ax=axs[0][0])
```

### Creation of histogram

```python
# We apply Sturges formula. See at https://www.lifeder.com/regla-sturges/
bins = int(np.ceil(1 + np.log2(df.shape[0])))

plt.hist(df["Day Charge"], bins=bins)
```

### Boxplots

```python
plt.boxplot(df["Day Calls"])
plt.ylabel("Frequency of day calls")
plt.title("Boxplot of day calls column")

# Compute interquartile range
iqr = df["Day Calls"].quantile(0.75) - df["Day Calls"].quantile(0.25)
# Maximum
df["Day Calls"].quantile(0.75) + iqr*1.5
# Minimum
df["Day Calls"].quantile(0.25) - iqr*1.5
```
